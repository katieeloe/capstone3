```python
rows = np.zeros(shape=(len(recipes_json),len(bool_feature_cols)))
df = pd.DataFrame(rows,columns=bool_feature_cols)
```

```python
df.head()
```

```python
cuisine_group = []
all_cuisines = []
for idx, recipe in enumerate(recipes_json):
    cuisine1 = []
    for cuisine in recipe['cuisines']:
        df.iloc[idx][feature_dict[cuisine]] = 1
        cuisine1.append(feature_dict[cuisine])
        all_cuisines.append(feature_dict[cuisine])
    cuisine_group.append(" & ".join(sorted(cuisine1)))
```

```python
x = set(all_cuisines)
x
```

```python
cuisine_group_counts = pd.Series(cuisine_group)
cuisine_group_counts = cuisine_group_counts.apply(lambda x: "no_cuisine" if x == "" else x)
cuisine_group_counts = cuisine_group_counts.value_counts().reset_index()
cuisine_group_counts.columns = ['cuisine', 'count']
```

```python
dish_group = []
all_dt = []
for idx, recipe in enumerate(recipes_json):
    group = []
    for dt in recipe['dishTypes']:
        df.iloc[idx][feature_dict[dt]] = 1
        group.append(feature_dict[dt])
        all_dt.append(feature_dict[dt])
    dish_group.append(" & ".join(sorted(set(group))))
```

```python
y = set(all_dt)
y
```

```python
dish_group_counts = pd.Series(dish_group)
dish_group_counts = dish_group_counts.apply(lambda x: "no_dish_type" if x == "" else x)
dish_group_counts = dish_group_counts.value_counts().reset_index()
dish_group_counts.columns = ['dish_type', 'count']
dish_group_counts.head()
```

```python
diet_group = []
all_diets = []
for idx, recipe in enumerate(recipes_json):
    diets = []
    for diet in recipe['diets']:
        df.iloc[idx][feature_dict[diet]] = 1
        diets.append(feature_dict[diet])
        all_diets.append(feature_dict[diet])
    diet_group.append(" & ".join(sorted(diets)))
```

```python
diet_group_counts = pd.Series(diet_group)
diet_group_counts = diet_group_counts.apply(lambda x: "no_diet" if x == "" else x)
diet_group_counts = diet_group_counts.value_counts().reset_index()
diet_group_counts.columns = ['diet', 'count']
diet_group_counts.head()
```

```python
simplified_ingredients = []
combined_ingredients_2 = []
unbucketed_ingredients = []
original_ingredients = []

for recipe in recipes_json:
    _ = []
    for ingredient in recipe['extendedIngredients']:
        original_ingredients.append(ingredient['name'].lower())
        if ingredient['name'].lower() in remove_set:
            continue
        elif ingredient_dict.get(ingredient['name'].lower()) == None:
            unbucketed_ingredients.append(ingredient['name'].lower())
        else:
            _.append(ingredient_dict.get(ingredient['name'].lower()))
            combined_ingredients_2.append(ingredient_dict.get(ingredient['name'].lower()))
    simplified_ingredients.append(_)
```
